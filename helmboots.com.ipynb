```python
import requests
```

```python
url = "https://helmboots.com/products.json?limit=250"
data = requests.get(url).json()
```

```python
products_list = []
for boot in data['products']:
    product_id = boot['id']
    title = boot['title']
    handle = boot['handle']
    created_at = boot['created_at']
    vendor = boot['vendor']
    product_type = boot['product_type']
    for image in boot['images']:
        try:
            image = image['src']
        except:
            image = "None"
    for variant in boot['variants']:
        available = variant['available']
        price = "$"+variant['price']
        weight = str(variant['grams'])+"g"
        sku = variant['sku']
        shipping = variant['requires_shipping']
        Tax = variant['taxable']
        product_data = {
            "product_id":product_id,
            "title":title,
            "handle":handle,
            "created_at":created_at,
            "vendor":vendor,
            "product_type":product_type,
            "available":available,
            "price":price,
            "weight":weight,
            "sku":sku,
            "shipping":shipping,
            "tax":Tax,
            "image":image
        }
        products_list.append(product_data)
print(products_list)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(products_list)
```

```python
df.set_index('product_id',inplace=True)
```

```python
df.to_csv("helmboots.csv")
```
